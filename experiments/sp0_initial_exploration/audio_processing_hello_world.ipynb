### Loading libraries

```python
import pretty_midi
import note_seq
from note_seq.midi_io import midi_to_note_sequence, note_sequence_to_pretty_midi

import IPython.display
```

### Loading one midi song & exploring

```python
# Load MIDI file into PrettyMIDI object. Give path to a midi file
midi_data = pretty_midi.PrettyMIDI('midi/0.mid')
```

```python
#Shows instruments. Sometimes Instruments will have names.
midi_data.instruments
```

```python
# One of the ways to play song is by using IPython. 
fs = 44100 # sample rate. It is like frames in the videos, showing how many samples are taken within a second from an audio, in kHz.
IPython.display.Audio(midi_data.synthesize(fs=fs), rate=fs)
```

```python
# Another way is to use NoteSequence object to achieve it from note_seq library. Conversion is easy:
sequence = midi_to_note_sequence(midi_data)
```

```python
# Playing song:
note_seq.play_sequence(sequence)
```

```python
# note_seq library has a way to plot the pianoroll easily as well:
note_seq.plot_sequence(sequence)
```
